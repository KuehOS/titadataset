```python
#Ages

def freq_age_groups(dat1):
    age_0_20 = 0
    age_20_60 = 0 
    age_60 = 0
    for ages in dat1["age"]:
        if ages <= 20:
            age_0_20 += 1
        elif 20 < ages <= 60:
            age_20_60 += 1
        else:
            age_60 += 1
    age_list = [age_0_20, age_20_60 , age_60]        
    return age_list


    
freq_age_groups(df)

plt.pie(freq_age_groups(df), labels = ["Aged 0 to 20","Aged 20 to 60" ,"Age over 60"])
plt.axis("equal")
plt.show() 
        
        
        
```
